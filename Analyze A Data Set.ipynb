This is Boston Housing Data from a CSV file. Dataset: https://archive.ics.uci.edu/ml/datasets/Housing

Table column descriptions:
    
    1. CRIM      per capita crime rate by town
    2. ZN        proportion of residential land zoned for lots over 
                 25,000 sq.ft.
    3. INDUS     proportion of non-retail business acres per town
    4. CHAS      Charles River dummy variable (= 1 if tract bounds 
                 river; 0 otherwise)
    5. NOX       nitric oxides concentration (parts per 10 million)
    6. RM        average number of rooms per dwelling
    7. AGE       proportion of owner-occupied units built prior to 1940
    8. DIS       weighted distances to five Boston employment centres
    9. RAD       index of accessibility to radial highways
    10. TAX      full-value property-tax rate per $10,000
    11. PTRATIO  pupil-teacher ratio by town
    12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks 
                 by town
    13. LSTAT    % lower status of the population
    14. MEDV     Median value of owner-occupied homes in $1000's

**QUESTION: Is there a relationship between the proportion of residentail land zoned for lots over 25,000 sq.ft. and the per capita crime rate by town?**

```python
x = df["ZN"]
```

```python
y = df["CRIM"]
```

```python
plt.plot(x, y, 'ro')
plt.show()
```

Detail:

```python
plt.plot(x, y, 'ro')
plt.axis([0, 10, 0, 10])
plt.show()
```

Detail 2:

```python
plt.plot(x, y, 'ro')
plt.axis([10, 40, 0, 2])
plt.show()
```

**ANSWER: It seems that there are plenty of neighborhoods that have no residential land zoned for lots over 25,000 sq.ft. and that many of these neighborhoods have very high crime rates relative to all neighborhoods that have residential land zoned for lots over 25,000 sq.ft. Indeed, the only high crime rates are observed in neighborhoods where there is no land zoned for lots over 25,000 sq.ft. There thus seems to be some kind of correlation between, in effect, population density and crime rates. However, there is an interesting spike in crime rates in neighborhoods where the percentage of land zoned for lots over 25,000 sq.ft. is around 20. Crime seems to spike at this level of population density at the time of the collection of this data.**
